```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
data_dir = 'dataset/'
forged_dir = os.path.join(data_dir, 'forged')
authentic_dir = os.path.join(data_dir, 'authentic')

authentic_dir
```

```python
forged_images = [os.path.join(forged_dir, f) for f in os.listdir(forged_dir) if f.endswith('.png')]
authentic_images = [os.path.join(authentic_dir, f) for f in os.listdir(authentic_dir) if f.endswith('.jpg')]
forged_images
```

```python
train_forged, test_forged = train_test_split(forged_images, test_size=0.2, train_size=0.8, random_state=42)
train_authentic, test_authentic = train_test_split(authentic_images, test_size=0.2, train_size=0.8, random_state=42)
```

```python
train_dir = 'train/'
test_dir = 'test/'

for directory in [train_dir, test_dir]:
    for sub_dir in ['forged', 'authentic']:
        os.makedirs(os.path.join(directory, sub_dir), exist_ok=True)
```

```python
def move_files(file_list, target_dir):
    for file in file_list:
        shutil.copy(file, target_dir)

move_files(train_forged, os.path.join(train_dir, 'forged'))
move_files(train_authentic, os.path.join(train_dir, 'authentic'))
move_files(test_forged, os.path.join(test_dir, 'forged'))
move_files(test_authentic, os.path.join(test_dir, 'authentic'))
```
