```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
data_dir = 'dataset/'
forged_dir = os.path.join(data_dir, 'forged')
authentic_dir = os.path.join(data_dir, 'authentic')

authentic_dir
```

```python
forged_images = [os.path.join(forged_dir, f) for f in os.listdir(forged_dir) if f.endswith('.png')]
authentic_images = [os.path.join(authentic_dir, f) for f in os.listdir(authentic_dir) if f.endswith('.jpg')]
forged_images
```

```python
def analyze_images(image_paths, label):
    image_shapes = []
    for img_path in image_paths:
        img = cv2.imread(img_path)
        if img is not None:
            image_shapes.append(img.shape)
    return image_shapes
```

```python
forged_shapes = analyze_images(forged_images, 'forged')
authentic_shapes = analyze_images(authentic_images, 'authentic')

forged_shapes = np.array(forged_shapes)
authentic_shapes = np.array(authentic_shapes)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(forged_shapes[:, 0], bins=20, alpha=0.5, label='Height')
plt.hist(forged_shapes[:, 1], bins=20, alpha=0.5, label='Width')
plt.legend()
plt.title('Forged Images Dimensions')

plt.subplot(1, 2, 2)
plt.hist(authentic_shapes[:, 0], bins=20, alpha=0.5, label='Height')
plt.hist(authentic_shapes[:, 1], bins=20, alpha=0.5, label='Width')
plt.legend()
plt.title('Authentic Images Dimensions')

plt.show()
```

```python
labels = ['forged'] * len(forged_images) + ['authentic'] * len(authentic_images)
sns.countplot(labels)
plt.title('Distribution of Labels in the Dataset')
plt.show()
```
